## Crear tablas

Aquí inserta el código SQL que creara las tablas de la base de datos:

### Tabla "Usuarios"

```python
%%sql
CREATE TABLE Usuarios(
    ID_usuario INTEGER GENERATED BY DEFAULT AS IDENTITY,
    nombre VARCHAR (45) NOT NULL,
    apellido VARCHAR (45),
    correo VARCHAR (45) NOT NULL,
    contraseña VARCHAR (45) NOT NULL,
    fecha_registro TIMESTAMP NOT NULL,
    PRIMARY KEY (ID_usuario)
    
);
```

### Tabla "Creadores"

```python
%%sql
CREATE TABLE Creadores (
    ID_creador INTEGER GENERATED BY DEFAULT AS IDENTITY,
    nombre VARCHAR (45) NOT NULL,
    apellido VARCHAR (45) NOT NULL,
    popularidad INTEGER,
    Estilo VARCHAR (45),
    PRIMARY KEY (ID_creador)
    
);
```

### Tabla "Tipos_NTF"

```python
%%sql
CREATE TABLE Tipos_NFT (
    ID_tipo INTEGER GENERATED BY DEFAULT AS IDENTITY,
    nombre VARCHAR (20) NOT NULL,
    PRIMARY KEY (ID_tipo)


);
```

### Tabla "NFT"

```python
%%sql
CREATE TABLE NFT(
    ID_nft INTEGER GENERATED BY DEFAULT AS IDENTITY,
    nombre VARCHAR (45) NOT NULL,
    ID_creador INTEGER,
    ID_tipo INTEGER,
    PRIMARY KEY (ID_nft),
    FOREIGN KEY (ID_creador) REFERENCES Creadores(ID_creador),
    FOREIGN KEY (ID_tipo) REFERENCES Tipos_NFT(ID_tipo)
    


);
```

### Tabla "Precio NFT"

```python
%%sql
CREATE TABLE Precio_NFT (
    ID_nft INTEGER GENERATED BY DEFAULT AS IDENTITY,
    fecha TIMESTAMP NOT NULL,
    precio INTEGER NOT NULL,
    PRIMARY KEY (ID_nft, fecha),
    FOREIGN KEY (ID_nft) REFERENCES NFT(ID_nft)
    


);
```

### Tabla "Usuario tiene NFT"

```python
%%sql
CREATE TABLE Usuario_tiene_NFT(
    id_usuario INTEGER GENERATED BY DEFAULT AS IDENTITY,
    id_nft INTEGER GENERATED BY DEFAULT AS IDENTITY,
    PRIMARY KEY (id_usuario, id_nft),
    FOREIGN KEY (id_usuario) REFERENCES Usuarios(ID_usuario),
    FOREIGN KEY (id_nft) REFERENCES NFT(ID_nft)


);
```

### Consulta 3
**El nombre y tipo (nombre del tipo) del NFT históricamente más caro.**
###### Si hay más de uno, ponga solamente el primero
Ejemplo:

| NFT    | Tipo | 
|----------------|--------|
| Minecraft Sphere| Arte 3D |

Se utiliza NATURAL JOIN para obtener el nombre y tipo de NFT de la tablas NFT y Precio_NFT. Con la funcion MAX() dentro de WHERE se encuentra el precio más alto y con nft.id_tipo = tipos_nft.id_tipo se obtiene el nombre del tipo.

```python
%%sql
SELECT
    nft.nombre as "NFT", tipos_nft.nombre as "Tipo"
FROM
    NFT NATURAL JOIN Precio_NFT, Tipos_NFT
WHERE precio = (SELECT MAX(precio) FROM Precio_NFT) 
        and nft.id_tipo = tipos_nft.id_tipo
```

```python
%%sql
SELECT nombre as NFT
    
FROM NFT 
NATURAL JOIN precio_nft
WHERE 
    id_tipo = 1 and fecha = (SELECT max(fecha) FROM precio_nft WHERE precio_nft.id_nft = nft.id_nft)
GROUP BY nombre, precio
ORDER BY precio DESC
LIMIT 4



```

Se hace uso de NAT. JOIN para que COUNT acceda a la tabla usuario_tiene_nft. Luego estos datos se clasifican con GROUP, ORDER y LIMIT para obtener al creador con más ventas

```python
%%sql
SELECT
    nombre as "Nombre", apellido as "Apellido", COUNT(id_NFT) as "Cantidad"
FROM 
    Creadores NATURAL JOIN usuario_tiene_nft
GROUP BY nombre, apellido, id_nft
ORDER BY COUNT(id_NFT) DESC
LIMIT 1
```
